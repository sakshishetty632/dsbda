```python
html=bs4.BeautifulSoup(req.text)
print(html)
```

```python
reviews = html.findAll('div',{'class':'t-ZTKy'});
for review in reviews :
    print(review.get_text()+"\n")
```

```python
ratings = html.findAll('div',{'class':'_3LWZlK _1BLPMq'});
for rating in ratings :
    print(rating.get_text()+"\n")
```

```python
qas = html.findAll('div',{'class':'_1RWRBu'});
for qa in qas :
    print(qa.get_text()+"\n")
```

```python
custs = html.findAll('p',{'class':'_2sc7ZR _2V5EHH'});
for cust in custs :
    print(cust.get_text()+"\n")
```

```python
import csv
```

```python
with open('scrapped_data.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)

    # Write headers to the CSV file
    writer.writerow(['Review', 'Rating', 'QA', 'Customer'])

    # Iterate over the data and write each entry to a row in the CSV file
    for i in range(max(len(reviews), len(ratings), len(qas), len(custs))):
        review = reviews[i].get_text() if i < len(reviews) else ""
        rating = ratings[i].get_text() if i < len(ratings) else ""
        qa = qas[i].get_text() if i < len(qas) else ""
        customer = custs[i].get_text() if i < len(custs) else ""
        writer.writerow([review, rating, qa, customer])

print("Data successfully imported into scrapped_data.csv")
```
