# Creating Subsets

```python
subset1 = df[['Type','Category','like','share']].loc[:150]
```

```python
subset2 = df[['Type','Category','like','share']].loc[151:300]
```

```python
subset1
```

```python
subset2
```

# Merging

```python
df_merged = pd.merge(subset1,subset2,how='outer')
```

```python
df_merged
```
